```python
import matplotlib
from matplotlib import pyplot as plt
```

```python
student=["Priya","Rohan","Sunil","Aashika"]
mark=[10,30,20,40]
plt.plot(student,mark,color="g",ls=":",label="Marks",lw=10)
plt.title("Marks")
plt.legend()
plt.xlabel("Students")
plt.ylabel("Marks")
```

```python
colr=['r','y','k','m']
plt.bar(student,mark,align="center",width=[1.4,1.7,1.2,1],color=["y","b","r","g"],ec="b",label="Marks",ls="--")
plt.grid(axis="x",linestyle=":",lw=2,color="red")
plt.legend()
plt.title("Marks")
plt.xlabel("Students")
plt.ylabel("Marks")
```

```python
plt.scatter(student,mark,s=200,marker="p",color="r",ec="b",lw=1)
```

```python
import matplotlib.pyplot as plt
stu=["A","B","C","D","E"]
ai_mrk=[60,70,80,90,50]
it_mrk=[50,30,40,20,10]
plt.plot(stu,ai_mrk,'r''h:',label="AI")
plt.plot(stu,it_mrk,'g''H:',label="IT")
plt.title("marks comparision")
plt.legend()
plt.xlabel("students")
plt.ylabel("score")

plt.show()
```

```python
#simple vertical bar plot
#ec=edgecolor
name=["raj","dev","krina","suhani"]
percentage=[50,80,60,90]
colr=['r','y','k','m']
wid=[0.2,0.3,0.4,0.5]
plt.bar(name,percentage,ls="--",align="center",width=wid,color=colr,ec="b",lw=2,label="percentage")
#color='r' all bars will appear red. width=0.2 all bars will be of same size
#color parameter can have hexa codes #ff00ff or color name "blue"
#align =edge/center will display x-axis names in left/center of the bar
#linestyle=:,-,-.,.,
#alpha =0.5 , will change color transparency. value can be from 0 to 1
plt.grid(axis='x',linestyle=':',linewidth=2, color="red")
plt.legend(loc="center right")
#loc=1,2,3,4,upper right,lower right, upper left, lower left, center, center right, center left
plt.title("CLASS X RESULT")   #fontsize=15
plt.xlabel("students name")
plt.ylabel("percentage")
plt.show()
```

```python
#simple horizontal bar plot
#ec=edgecolor
name=["raj","dev","krina","suhani"]
percentage=[50,80,60,90]
wid=[0.2,0.3,0.4,0.5]
plt.bar(name,percentage,linestyle="-.",color="r",ec="black", label="percentage")
plt.grid()
plt.legend(loc="center right")
plt.title("CLASS X RESULT")
plt.xlabel("students name")
plt.ylabel("percentage")
plt.show()
```

```python
#comparision using bar plot
#xticks and yticks is used to change the axis sequence
name=["raj","dev","krina","suhani"]
english=[50,80,60,90]
hindi=[90,40,80,30]
width=0.2
p=np.arange(len(name))
p1=[j+width for j in p]
plt.bar(p,english,width,color='r',label="english",align="edge")
plt.bar(p1,hindi,width,color='b',label="hindi",align="edge")
plt.xticks(p+width,["raj","dev","krina","suhani"])
plt.grid()
plt.legend(loc="center right")
plt.title("CLASS X RESULT")
plt.xlabel("students name")
plt.ylabel("Marks Score")
plt.show()
```

```python
#scatter plot
#s-square, o-circle, d-diamond,p-pentagon,h-hexagon,^-triangle,
name=["raj","dev","krina","suhani"]
english=[50,80,60,90]
hindi=[90,40,80,30]
plt.scatter(name,english,s=400,marker="p",color="red",ec="blue",linewidth=2,alpha=0.6)
plt.scatter(name,hindi,s=400,marker="p",color="blue",ec="black",linewidth=2,alpha=0.2)
plt.show()
```

```python
#histogram plot
#histtype - bar,step,barstacked,stepfilled. width parameter is req only with bar option.
Freq=[1,3,5,7,7,8,9,9,10,10,11,12,13,14,15]
pl=[0,3,6,9,12,15]
plt.hist(Freq,5,(0,15),histtype="bar",width=0.6,align="right") #bins means groups or bars that we see on plot
#0-15 shows a-axis value
plt.xticks([3,6,9,12,15])
plt.show()
```

```python
#bins means group of range 0-4,5-9,10-15
#histogram using random integer value
x=np.random.randint(1,15,(15)) # from 1 to 14 range, 15 elements
print(x)
plt.hist(x,3,(0,15),histtype="bar",width=0.5,align="right",ec="r")
plt.xticks([5,10,15])
plt.show()
```

```python
#pie plot
x=[10,20,30,40]
y=["c","c++","java","python"]
ex=[0.2,0.0,0.4,0.0]
plt.pie(x,explode=ex,labels=y,radius=1.5,startangle=90,autopct="%0.1f%%",counterclock=False)
plt.show()
```

```python
#x/y limit and x/y ticks dont give you desired axis names
x=["jan","feb","march","april"]
y=[300,200,400,100]
plt.xlim([0,3])
plt.ylim([100,500])
plt.plot(x,y)
plt.show()
```

```python
#subplot
student=["a","b","c","d"]
marks=[34,45,67,89]
plt.subplot(2,2,1)
plt.plot(student,marks,label= "totalmarks")
plt.legend()
#-----------------
plt.subplot(2,2,2)
plt.bar(student,marks,label= "totalmarks")
plt.yticks(marks,[0-20,20-40,40-60,60-80])
plt.legend()
#---------------
plt.subplot(2,2,3)
plt.barh(marks,student,height=10)
plt.legend()
#----------------
plt.subplot(2,2,4)
plt.scatter(student,marks, marker="h", s=30,label= "totalmarks")
plt.suptitle("four types of plotting")
plt.legend()
plt.show()
```
